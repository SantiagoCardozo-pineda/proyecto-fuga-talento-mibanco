Creamos un codigo que define grupos de edad, se analiza la distribucion de las categorias y se crea una funcion para ya poner el estado civil basados en la distribucion y el grupo de edad

```python
# Definimos los grupos de edad
bins = [18, 30, 40, 50, 100]
labels = ['Joven', 'Adulto', 'Maduro', 'Mayor']
base_final['Grupo_Edad'] = pd.cut(base_final['N_EDAD'], bins=bins, labels=labels, right=False)

# Analizamos la distribución de estado civil en cada grupo de edad
distribucion_est_civil = base_final.groupby('Grupo_Edad')['S_EST_CIVIL'].value_counts(normalize=True).unstack().fillna(0)

# Función para imputar estado civil basado en la distribución y el grupo de edad
def imputar_estado_civil(grupo_edad, distribucion):
    if pd.isna(grupo_edad):
        return np.nan
    return np.random.choice(distribucion.columns, p=distribucion.loc[grupo_edad].values)

# Imputación de valores faltantes en la columna 'S_EST_CIVIL'
base_final['S_EST_CIVIL'] = base_final.apply(
    lambda row: imputar_estado_civil(row['Grupo_Edad'], distribucion_est_civil) if pd.isna(row['S_EST_CIVIL']) else row['S_EST_CIVIL'],
    axis=1
)

# Eliminamos la columna 'Grupo_Edad' ya que no queremos columnas adicionales
base_final.drop(columns=['Grupo_Edad'], inplace=True)

# Verificamos si aún existen nulos
nulos_restantes = base_final['S_EST_CIVIL'].isnull().sum()
```

```python
import re
from collections import Counter

# Diccionario de palabras clave y sus respectivas categorías
keywords = {
    "Motivos Personales": ["personal", "familia", "personales"],
    "Oferta de Trabajo": ["oferta", "sueldo", "Mejor Oportunidad Laboral"],
    "Motivos de Salud": ["salud"],
    "Desempeño": ["desempeño"],
    "Cambio de Ubicación": ["cambio", "residencia", "provincia"],
    "Pase a Planilla": ["planilla", "Pase a planilla", "PASE A PLANILLA"],
    "Tercerización": ["tercerización"],
    "Perfil": ["perfil"],
    "Renuncia Voluntaria": ["voluntaria", "renuncia"],
    "Problemas Familiares": ["problemas familiares"],
    "Mutuo Acuerdo": ["mutuo acuerdo"],
    "Reestructuración": ["reestructuración"]
}

# Función para categorizar observaciones
def categorize_observation(text):
    if pd.isna(text):  # Dejar los NaN tal como están
        return np.NaN

    text = text.lower()  # Convertir a minúsculas para uniformidad
    for category, words in keywords.items():
        if any(re.search(rf'\b{word.lower()}\b', text) for word in words):
            return category
    return "Otros"  # Si no coincide con ninguna categoría

# Aplicar la función de categorización
base_final_merge['MOTIVO_CESE'] = base_final_merge['S_OBS_CESE'].apply(categorize_observation)



# Mostrar la distribución de categorías
print(base_final_merge['MOTIVO_CESE'].value_counts(dropna=False))
```

### ** RENTABILIDAD DE EMPLEADOS **
Agregamos columna extra de rentabilidad de los empleados

```python
# Calculamos los años en la empresa
base_final_merge['FECHA_CONTRATACION'] = pd.to_datetime(base_final_merge['FECHA_CONTRATACION'])
base_final_merge['D_FEC_CESE'] = pd.to_datetime(base_final_merge['D_FEC_CESE'])
base_final_merge['TIEMPO_EMPRESA'] = base_final_merge.apply(lambda row: (row['D_FEC_CESE'] - row['FECHA_CONTRATACION']).days / 365
                                    if pd.notnull(row['D_FEC_CESE'])
                                    else (pd.Timestamp.today() - row['FECHA_CONTRATACION']).days / 365, axis=1)

# Costo inicial por adquisición
costo_adquisicion = 3317

# Calculamos el salario anual de cada empleado
base_final_merge['SALARIO_ANUAL'] = base_final_merge['RANGO_SALARIAL'] * 12

# Costo total para empleados activos
base_final_merge['COSTO_TOTAL'] = base_final_merge['RANGO_SALARIAL'] * base_final_merge['TIEMPO_EMPRESA'] + costo_adquisicion

# Para los empleados cesados, agregamos el costo de reemplazo
base_final_merge['COSTO_REEMPLAZO'] = base_final_merge['CESADO'].apply(lambda x: 3317 if x == 1 else 0)

# Costo total acumulado (incluye el costo de reemplazo si el empleado fue cesado)
base_final_merge['COSTO_TOTAL_ACUMULADO'] = base_final_merge['COSTO_TOTAL'] + base_final_merge['COSTO_REEMPLAZO']

# Rentabilidad (para todos, activos y cesados)
base_final_merge['RENTABILIDAD'] = (base_final_merge['RANGO_SALARIAL']-base_final_merge['RANG_SALARIAL_INICIAL'])*base_final_merge['TIEMPO_EMPRESA'] / base_final_merge['COSTO_TOTAL_ACUMULADO'] * 100

print(base_final_merge[['MATRICULA', 'RENTABILIDAD','CESADO']])
```

```python
#ELIMINAMOS LAS VARIABLES QUE NOS AYUDARON A OBTENER LA RENTABILIDAD
base_final_merge = base_final_merge.drop(columns=['COSTO_TOTAL','COSTO_REEMPLAZO','COSTO_TOTAL_ACUMULADO'])
base_final_merge.info()
```
