```python
import matplotlib.pyplot as plt
# Primero necesitamos codificar las variables categoricas a numericas para poder calcular la matriz de correlación
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

#La rentabilidad la consideramos aparte
RENTABILIDAD = data["RENTABILIDAD"]
data = data.drop("RENTABILIDAD", axis=1)

# Codificamos las variables categoricas
data_encoded = data.apply(le.fit_transform)

# Calculamos la matriz de correlación
correlation_matrix = data_encoded.drop(["PERIODO", "JORNADA_LABORAL"], axis=1).corr()

# Graficamos el mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlación')
plt.show()
```

```python
# Calculamos la matriz de correlación
correlation_matrix = data_encoded.drop(["PERIODO", "JORNADA_LABORAL", "EDAD"], axis=1).corr()

# Graficamos el mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlación')
plt.show()
```

Quitamos la edad ya que se encuentra correlacionada con el tiempo en el puesto y el grupo generacional asi que optamos solo por mantener el grupo generacional TOMANDO COMO PUNTO DE COORTE .3

Comenzamos con el modelado con una distribucion 80% y 20% utilizando una regresion logistica y un xgboost como modelos , utilizando el AUC como medida de presicion del modelo

```python

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

data_encoded['RENTABILIDAD'] = RENTABILIDAD
# Seleccionar las variables más importantes, incluyendo la rentabilidad para su posterior uso.
X = data_encoded[["RANGO_SALARIAL", "DIVISION", "PUESTO", "TIEMPO_PUESTO","RENTABILIDAD"]]
y = data_encoded['CESADO']

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

RENTABILIDAD = X_test["RENTABILIDAD"]
X_test = X_test.drop("RENTABILIDAD", axis=1)
X_train = X_train.drop("RENTABILIDAD", axis=1)

# Crear y entrenar el modelo de regresión logística
model = LogisticRegression()
model.fit(X_train, y_train)

# Predecir las probabilidades en el conjunto de prueba
y_pred_proba = model.predict_proba(X_test)[:, 1]

# Calcular el AUC
auc = roc_auc_score(y_test, y_pred_proba)
print("AUC:", auc)
```

```python
a = model.predict_proba(X_train)
a
```

```python
X_test["Prediccion"] = y_pred_proba
X_test["Target"] = y_test
X_test.head()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Predecir las probabilidades en el conjunto de entrenamiento
y_train_pred_proba = model.predict_proba(X_train)[:, 1]

# Calcular las curvas ROC para entrenamiento y prueba
fpr_train, tpr_train, _ = roc_curve(y_train, y_train_pred_proba)
roc_auc_train = auc(fpr_train, tpr_train)

fpr_test, tpr_test, _ = roc_curve(y_test, y_pred_proba)
roc_auc_test = auc(fpr_test, tpr_test)

# Graficar las curvas ROC
plt.figure()
plt.plot(fpr_train, tpr_train, color='darkorange', lw=2, label='ROC curve (train) (area = %0.2f)' % roc_auc_train)
plt.plot(fpr_test, tpr_test, color='navy', lw=2, label='ROC curve (test) (area = %0.2f)' % roc_auc_test)
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
